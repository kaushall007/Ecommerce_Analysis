```python
# Identify the most profitable products based on the difference between original and selling prices
```

```python

query = """
    SELECT 
        p.product_name, 
        p.original_price, 
        p.selling_price, 
        (p.selling_price - p.original_price) AS profit, 
        ((p.selling_price - p.original_price) / p.selling_price) * 100 AS profit_margin
    FROM 
        order_details od 
    JOIN 
        product p ON od.product_id = p.product_id
"""

# Execute the query and store the results in a DataFrame
cursor.execute(query)
result = cursor.fetchall()
df = pd.DataFrame(result, columns=['product_name', 'original_price', 'selling_price', 'profit', 'profit_margin'])

# Sort the DataFrame by profit margin in descending order
df_sorted = df.sort_values(by='profit_margin', ascending=False)

# Print the top 10 most profitable products
print(df_sorted.head(10))
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_sorted.head(10)['product_name'], df_sorted.head(10)['profit_margin'], color='skyblue')
plt.xlabel('Product')
plt.ylabel('Profit Margin (%)')
plt.title('Top 10 Most Profitable Products')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

# Customer Order Patterns:

```python
# Identify product names with the highest and lowest order quantities.
```

```python
query ="""
SELECT 
    p.product_Name, 
    SUM(od.quantity) AS quantity
FROM 
    order_details od
JOIN 
    order_details o ON od.order_id = o.order_id
JOIN 
    product p ON o.product_id = p.product_id
GROUP BY 
     p.product_Name
ORDER BY 
    quantity DESC;
    """
cursor.execute(query)
result = cursor.fetchall()
df = pd.DataFrame(result, columns=['product_name', 'quantity'])
df
```

```python

highest_order_product = df.loc[df['quantity'].idxmax()]
lowest_order_product = df.loc[df['quantity'].idxmin()]

plt.figure(figsize=(8, 6))
plt.bar([highest_order_product['product_name'], lowest_order_product['product_name']], 
        [highest_order_product['quantity'], lowest_order_product['quantity']])
plt.xlabel('product')
plt.ylabel('Quantity order')
plt.title('Highest and Lowest Order Quantities')
plt.xticks(rotation=0)  # No rotation needed for just two labels
plt.tight_layout()
plt.show()
```

```python
# Identify customers with the highest and lowest order quantities by customer name
```

```python
query ="""
SELECT 
    c.Name, 
    SUM(od.quantity) AS quantity
FROM 
    order_details od
JOIN 
    order_details o ON od.order_id = o.order_id
JOIN 
    customer c ON o.customer_id = c.customer_id
GROUP BY 
    c.Name
ORDER BY 
    quantity DESC;
    """
cursor.execute(query)
result = cursor.fetchall()
df = pd.DataFrame(result, columns=['name', 'quantity'])
df
```

```python

highest_quantity_customer = df.loc[df['quantity'].idxmax()]
lowest_quantity_customer = df.loc[df['quantity'].idxmin()]

plt.figure(figsize=(8, 6))
plt.bar([highest_quantity_customer['name'], lowest_quantity_customer['name']], 
        [highest_quantity_customer['quantity'], lowest_quantity_customer['quantity']])
plt.xlabel('Customer Name')
plt.ylabel('Quantity')
plt.title('Highest and Lowest Order Quantities')
plt.xticks(rotation=0)  # No rotation needed for just two labels
plt.tight_layout()
plt.show()
```

```python
# Analyze whether certain products or categories are more popular in a specific city
```

```python
query = """
    SELECT 
        c.city, 
        p.category, 
        COUNT(*) AS total_orders, 
        SUM(od.total_price) AS total_revenue
    FROM 
        order_details od 
    JOIN 
        order_details o ON od.order_id = o.order_id
    JOIN 
        customer c ON o.customer_id = c.customer_id
    JOIN 
        product p ON od.product_id = p.product_id
    GROUP BY 
        c.city, 
        p.category
    ORDER BY 
        c.city, 
        total_orders DESC
"""


# Execute the query and store the results in a DataFrame
cursor.execute(query)
result = cursor.fetchall()
df = pd.DataFrame(result, columns=['city', 'category', 'total_orders', 'total_revenue'])

# Print the results
df
```

```python

data = {
    'city': df['city'],
    'category': df['category'],
    'total_orders': df['total_orders'],
    'total_revenue': df['total_revenue']
}

df = pd.DataFrame(data)

# Pivot the data
pivot_table = df.pivot_table(values='total_orders', index='city', columns='category', fill_value=0)

# Plot the heatmap
plt.figure(figsize=(10, 6))
plt.imshow(pivot_table, cmap='viridis', aspect='auto')
plt.colorbar(label='Total Orders')
plt.xticks(ticks=np.arange(len(pivot_table.columns)), labels=pivot_table.columns)
plt.yticks(ticks=np.arange(len(pivot_table.index)), labels=pivot_table.index)
plt.title('Category Popularity in Each Region and City')
plt.xlabel('Category')
plt.ylabel('City')
plt.show()
```

# Time Analysis

```python
query = """
SELECT 
    MONTH(order_date) AS month,
    SUM(quantity) AS total_sales
FROM 
    order_details
GROUP BY 
    MONTH(order_date)
ORDER BY 
    month;
"""
cursor.execute(query)
result = cursor.fetchall()

df = pd.DataFrame(result,columns = ['months','sales'])

month_names = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}

df['months'] = df['months'].map(month_names)
df
```

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['months'], df['sales'], color='skyblue')
plt.xlabel('Months')
plt.ylabel('Total Sales ')
plt.title('Order Status Distribution')
plt.tight_layout()
plt.show()
```
